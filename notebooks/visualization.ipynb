```python
import sys
import os
import json
sys.path.append("/coc/pskynet6/dhe83/mice")
from src import config
```

```python
inference_dir = "/nethome/dhe83/mice/experiments/BoolQ/id_2_train_16_test_64/generations/similar_2_64_all-roberta-large-v1/2.7b"
```

```python
def read_json(filepath: str) -> dict:
    with open(filepath, "r") as f:
        return json.load(f)
```

```python
def get_examples(inference_dir)
    examples = {}
    for f in os.listdir(inference_dir):
        f_path = os.path.join(inference_dir, f)
        if os.path.isdir(f_path):
            mv_path = os.path.join(f_path, "mice-sampling_sampled_probs.json")
            mice_path = os.path.join(f_path, "mice-sampling_combined_probs.json")

            mv = read_json(mv_path)
            mice = read_json(mice_path)

            examples[os.path.basename(f)] = {"mv": mv, "mice": mice}
    return examples

examples = get_examples(inference_dir)
```

```python
def get_probs(examples):
    mv_probs = {}
    mice_probs = {}
    for k, v in examples.items():
        mv_data = v['mv']
        gold_prompts = 0
        total_prompts = 0
        for span in mv_data:
            if span['is_gold']:
                gold_prompts+= span['num_prompts']
            total_prompts+=span['num_prompts']

        mv_probs[k] = float(gold_prompts) / total_prompts

        mice_data = v['mice']
        l = [x for x in mice_data if x['is_gold']]
        mice_probs[k] = l[0]['prob'] if len(l) > 0 else 0
    return {"mv_probs": mv_probs, "mice_probs": mice_probs}

probs = get_probs(examples)
mv_probs = probs['mv_probs']
mice_probs = probs['mice_probs']
```

```python
import matplotlib.pyplot as plt
# the histogram of the data

def plot_probs(probs: list, desc: str):
    n, bins, patches = plt.hist(probs, 50, density=True, facecolor='g', alpha=0.75)

    plt.xlabel('Probability')
    plt.ylabel('Frequency')
    plt.title('Histogram of Gold Probabilities')
    #plt.text(0.2, 8, 'Majority Vote on BoolQ')
    plt.axis([0, 1, 0, 10])
    plt.grid(True)
    plt.show()
    
plot_probs(list(mv_probs.values()), "Majority Vote for BoolQ")
plot_probs(list(mice_probs.values()), "MICE Sampling for BoolQ")
```

```python
def verbalize(pred):
    return pred.lower().strip() in ["yes", "true"]
```

```python
in_context_gold = {}
in_context_total = {}

for f in os.listdir(inference_dir):
    f_path = os.path.join(inference_dir, f)
    if os.path.isdir(f_path):
        predictions_path = os.path.join(f_path, "predictions.json")

        predictions = read_json(predictions_path)
        mice = read_json(mice_path)
        
        for prompt, v in predictions.items():
            if v['label'] == verbalize(v['prediction']):
                 in_context_gold[prompt] = in_context_gold.get(prompt, 0) + 1   
            in_context_total[prompt] = in_context_total.get(prompt, 0) + 1

            
in_context_accuracy = {}
for k, v in in_context_total.items():
    in_context_accuracy[k] = float(in_context_gold[k] if k in in_context_gold else 0) / in_context_total[k]

print(in_context_accuracy)
```

```python
from ast import literal_eval as make_tuple

demonstrations = set()
for k in in_context_gold.keys():
    for x in make_tuple(k):
        demonstrations.add(x)
        
stoi = {s:i for i, s in enumerate(demonstrations)}
itos = {s:i for i, s in stoi.items()}
```

```python
print(itos)
```

```python
import torch
import numpy as np
N = len(demonstrations)
grid = torch.zeros((N, N))

for k, v in in_context_accuracy.items():
    ids = make_tuple(k)
    grid[stoi[ids[1]]][stoi[ids[0]]] = v
           

fig, ax = plt.subplots()
im = ax.imshow(grid, cmap="Blues")

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(demonstrations)), labels=demonstrations)
ax.set_yticks(np.arange(len(demonstrations)), labels=demonstrations)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")



ax.set_title("Accuracy of prompt combinations")
fig.tight_layout()
plt.show()
```
